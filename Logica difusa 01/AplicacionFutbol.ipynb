```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(30,80,0.1)

lento = fuzz.trimf(x,[30,30,50])
medio = fuzz.trimf(x,[30,50,70])
medio_rapido = fuzz.trimf(x,[50,60,80])
rapido = fuzz.trimf(x,[60,80,80])

plt.figure()
plt.plot(x,lento,'b',linewidth=1.5,label='Lento')
plt.plot(x,medio,'k',linewidth=1.5,label='Medio')
plt.plot(x,medio_rapido,'m',linewidth=1.5,label='Medio-Rapido')
plt.plot(x,rapido,'r',linewidth=1.5,label='Rapido')

plt.title('Penalti Difuso')
plt.ylabel('Membresía')
plt.xlabel('Velocidad km/h')
plt.legend(loc='center right',bbox_to_anchor=(1.25,0.5),ncol=1,fancybox=True,shadow=True)
```
